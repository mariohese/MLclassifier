```python
from preprocessor import preProcessSerie
import json
import time
import requests
import pandas as pd
    
def getSentiment(tweet, lista, counter):
    response = ""
    url = "http://api.meaningcloud.com/sentiment-2.1"
    tweet = tweet.encode('latin-1', 'ignore')
    tweet = tweet.decode(encoding='latin-1')
    tweet = tweet.encode('utf-8')

    apikeys = ['adb0e1dc1e6d7cf39d85af3f9d22d7b7', 'f14c341f5d79d1a292a9ba2addca0a99', '2773f584014b5f3183a638f71ce58e47']
    apikey = ''
    if (counter >= 18000 and counter<=37000):
        apikey = apikeys[1]
    elif counter > 37000:
        apikey = apikeys[2]
    else:
        apikey = apikeys[0]
    
    payload = "key="+apikey+"&lang=en&txt="+str(tweet,'utf-8')
    
    headers = {'content-type': 'application/x-www-form-urlencoded'}
    response = requests.request("POST", url, data=payload, headers=headers)
        
    #for key, value in response.items() :
        #print (key)
    sentiment = ""
    try:
        sentiment = response.json()['score_tag']
    except KeyError:
        sentiment = "NONE"
        
    lista.append(sentiment)
        
    response.connection.close()
    time.sleep(1)
    if counter == 0:
        print(sentiment)
    if counter == 10:
        print(sentiment)
    counter= counter + 1
    if counter % 100 == 0:
        print(counter)
    return (lista, counter)
    
def sentiments(data, y=None):
    dataproc = preProcessSerie(data)
    counter = 0
    lista = []
    for tweet in dataproc:
        lista, counter = getSentiment(tweet, lista, counter)
    return lista
```
