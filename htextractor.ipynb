```python
import pandas as pd
import pickle
from preprocessor import preProcessAllText
from nltk.tokenize import TweetTokenizer



tknzrwhu = TweetTokenizer(preserve_case=False, reduce_len=True, strip_handles = False)


def notinlist(item, lista):
    booleano = True
    if item in lista:
        booleano = False
    return booleano 

def allHashtags(tweet, lista):
    sent = tknzrwhu.tokenize(str(tweet) )    
    for term in sent:
        if term.startswith('#'):
            if notinlist(term, lista):
                lista.append(term)
    return lista

def transform(data):
    print("Entra en hashtags")
    dataproc = preProcessAllText(data)
    lista = []
    lista = allHashtags(dataproc, lista)
    
    return lista


```

```python
import pickle

isis = pd.read_csv('isisfanboy.csv')
about = pd.read_csv('aboutisis.csv')

isis = isis[:17392]
about = about[:17392]
#El máximo que puedo usar es 17392

dataframe = pd.concat([isis, about])
X1 = isis['tweets'].values

X2 = about['tweets'].values

listahtisis = transform(X1)
print("Ha acabado con los del isis")
listahtabout = transform(X2)
```

```python
listaht = listahtisis + listahtabout
listaht = list(set(listaht))
```

```python
pickle.dump(listaht, open( "ht.p", "wb" ) )
```
